# definiendo metodos 


```python
class penguin :
    #definicio de atributo
    categoria = "AVE"
    def __init__(self, nombre, especie):
        print("creando pingiuno  :" + nombre + especie)
        # atribtos de instancia 
        self.nombre  =nombre
        self.especie = especie
        # definir metodos 
    def nada (self):
        print("woooooshhhhh")
    def camina (self , pasos ):
        print("caminando " + pasos + "paos , lenta y torpemente")
    
        #crenando un objeto 

mi_pinguino = penguin("santi","REY")

# llamada de metodos 
mi_pinguino.nada()
mi_pinguino.camina("200")


        
```
